```python
dataset = pd.read_csv("tictac_single.txt",sep=' ',names=['x1','x2','x3','x4','x5','x6','x7','x8','x9','y'])
```

```python
dataset.head()
```

```python
X = dataset.iloc[:,:-1]
```

```python
X.shape
```

```python
y = dataset.iloc[:,-1]
y.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
models = {
    'SVM': SVC(),
    'RandomForest':RandomForestClassifier(),
    'KNeighbors':KNeighborsClassifier(),
    'HGBoosting':GradientBoostingClassifier()
}

param_models = {
    'SVM':{
        'C': [0.01,0.1,1,5,10,100],
        'kernel':['linear', 'poly', 'rbf'],#, 'sigmoid', 'precomputed'],
        'gamma':['scale','auto',0.001,0.01,0.1,1],
        'coef0':[0.1,0.01,1,0.001]
    },
    'RandomForest':{
        'n_estimators':[10,100,500,1000],
        'max_depth':[5,10,50,None],
        'max_features':['sqrt','log2']
    },
    'KNeighbors':{
        'n_neighbors':[5,10,15,20,50],
        'weights':['uniform','distance'],
        'algorithm':['ball_tree','auto','kd_tree']
    },
    'HGBoosting':{
        'loss':['log_loss'],
        'learning_rate':[0.1,0.2,0.4,0.5,1],
        'n_estimators':[10,100,500,1000],
        'criterion':['friedman_mse','squared_error']

    }
}
```

```python
results = {}
for name,model in models.items():
  print('Now Training:',name)
  searchParam = RandomizedSearchCV(estimator=model,param_distributions=param_models[name],cv=5)

  searchParam.fit(X_train,y_train)

  results[name] = searchParam.best_estimator_
print(results)
```
